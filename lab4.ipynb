```python
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
sys.path.append('/content/drive/My Drive/Colab_Notebooks/COI')
from utility import util

image = cv.imread('/content/drive/My Drive/Colab_Notebooks/COI/V3JPHvAOsSU.jpg')
grayImage = cv.cvtColor(image, cv.COLOR_RGB2GRAY)

plt.figure(figsize=(32, 2))
plt.imshow(image, cmap='gray')
plt.title('Исходное изображение', fontsize=14)
plt.xticks([]), plt.yticks([])
```

```python
threshold = 115
ret1, thresh1 = cv.threshold(grayImage, threshold, 255, cv.THRESH_BINARY)

cv.imwrite('/content/drive/My Drive/Colab_Notebooks/COI/result.jpg', thresh1)

plt.figure(figsize=(32, 2))
plt.imshow(thresh1, cmap='gray')
plt.title('Бинаризация полутоновых изображений', fontsize=14)
plt.xticks([]), plt.yticks([])
```

```python
ret2, thresh2 = cv.threshold(grayImage, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

plt.figure(figsize=(32, 2))
plt.imshow(thresh2, cmap='gray')
plt.title('Бинаризация Оцу', fontsize=14)
plt.xticks([]), plt.yticks([])
```

```python
adaptive = cv.adaptiveThreshold(grayImage, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 5, 3)

plt.figure(figsize=(32, 2))
plt.imshow(adaptive, cmap='gray')
plt.title('Адаптивная бинаризация', fontsize=14)
plt.xticks([]), plt.yticks([])
```

```python
gaussian33 = cv.GaussianBlur(grayImage, (3, 3), 0)
gaussian55 = cv.GaussianBlur(grayImage, (5, 5), 0)

dog = gaussian33 - gaussian55


outImageDepth = cv.CV_16S
laplace = cv.Laplacian(grayImage, outImageDepth, ksize=1)

plt.figure(figsize=(32, 2))
plt.imshow(laplace, cmap='gray')
plt.title('Оператор Лапласа', fontsize=14)
plt.xticks([]), plt.yticks([])


plt.figure(figsize=(32, 2))
plt.imshow(dog, cmap='gray')
plt.title('Разность гауссовских распределений', fontsize=14)
plt.xticks([]), plt.yticks([])


log = cv.Laplacian(gaussian33, outImageDepth, ksize=1)
log = cv.convertScaleAbs(log)

plt.figure(figsize=(32, 2))
plt.imshow(log, cmap='gray')
plt.title('Лапласиан Гауссиана', fontsize=14)
plt.xticks([]), plt.yticks([])
```

```python
threshold1 = 100
threshold2 = 200
canny = cv.Canny(grayImage, threshold1, threshold2)

plt.figure(figsize=(32, 2))
plt.imshow(canny, cmap='gray')
plt.title('Оператор Канни', fontsize=14)
plt.xticks([]), plt.yticks([])

plt.show()
```
